## Part 2: Linear ClassifiersSo far in our class you learned how to design a binary classifier meaning that you only had two different labels (+1,-1). But here we have more than two categories (101 categories). How can we learn a classifier to classify examples from more than two classes?This problem is called "Multiple-Class Classification" (see CIML Chapter 5).To deal with this problem, we can learn separate binary classifiers for each category independently (e.i. 101 different classifiers) by taking the positive samples from one category and the negative samples from all the remaining categories.For a given test sample we apply all the 101 classifiers on it and choose the category label based on the classifier that gave us the most confident classification score. This method is called One-vs-All. <font color=#B0171F>Q3:Can you suggest any other way to deal with multiple class classification? Compare it with One-vs-All.</font>Response: First of all, we'd like to bring up something that was also mentioned in class, and it has to do with our doubts as to whether OVA is a recommended classificationscheme when the underlying classifiers are linear. To make things easier, let us make the simplifying assumption that all points have two features, thereforewe are classifying on the $\mathcal{R}^2$ space. With 101 classes, it seems unreasonable to assume that a linear classifier which would attempt to separate every class from every other class would do a good job. Now, if we generate this thought process into higher dimenions, where distances between objects becomeroughly the same (as discussed in the course and in CIML), the problem becomes more intense. It seems that you cannot possibly train an OVA classificationscheme with linear classifiers underneath without dealing with an obscene amount of False Positives. At this point, however, it should be said that Halmentions in CIML, page 73, that OVA works very well in practice as long as the hyper-parameters are tuned well. For the perceptron, the hyper-parameter $\textit{MaxIter}\ $ controls overfitting versus underfitting, with higher values of $\textit{MaxIter}\ $ indicating an effort to fit the training data better, possiblywith an adverse effect on generalization. In the soft-margin SVM, the hyper-parameter $\lambda$ or $\textit{C}$ also controls overfitting versus underfitting,because it controls the total amount of slack (misclassified points) we allow to our algorithm. So, in both cases, given the fundamental problem with OVA andlinear classifiers, I cannot see how Hal can be proven correct with respect to this set-up: if your algorithm's inductive bias is ill-suited for a particularlearning problem, then tuning and re-tuning its hyper-parameters won't help much.The way we understand it, the alternative of All-vs-All (AVA) classification is much preferred to OVA when the underlying classifiers are linear. It seems likea much better job to try to find a collection of $\frac{\mathcal{K ( K - 1)}}{2}$ hyperplanes that discern between pairs of classes, rather than having $\mathcal{K}$hyperplanes which try to separate each class from the rest of the dataset.<font color=#B0171F>Q4: In the One-vs-All senario you will have much more negative samples than positives for each classifier. Does this fact cause a reduction in the performance of the learning algorithm? if yes, why? and what is the your solution?</font>Response: The Perceptron is a perfect example on how imbalanced cardinalities between the positive and negative class can have a severe impact on the algorithm'srecognition accuracy. As discussed on CIML, chapter 3, when faced with imbalanced data, the Perceptron tends to stop learning from the minority class. Let us attemptto formulate this mathematically via an example. Suppose that we have a positive class with C points and we pit that class against a negative class with 100 * C points.We will now make two assumptions on the labels' distribution: One extreme, in the sense that it does not appear very often in practice, and one not so extreme, in the sense that it comescloser to reality.Assumption 1: The C minority class points are scanned first by the Perceptron, which then moves on to the 100*C points of the negative (majority) class.In this case, after a probably small number of points considered (e.g, 5 points), the Perceptron will stop learning and consider all points that come as positive. This is ok until the negative points arrive, in which case it will misclassify some points, and then gradually shift the decision boundary to accomodatethe negative points. Then, for a large number of points (somewhere in between 99 * C and 100 * C, the perceptron will once again learn nothing. Therefore, thedecision boundary for this case will not generalize well to test data becaause it will be either biased on the negative class (if the negative points are examinedlast) or the positive class (if they are examined last).Assumption 2: The C minority class points are distributed uniformly across the training data, exactly 100 negative points away from each other.We have discussed how important reshuffling of the examples is for the Perceptron. Assumption 2 can be derived from Assumption 1 after a $\textit{fair}\ $ reshuffling,where every point has the same probability of appearing at any order in the list of points examined. We can assume that the state of Assumption 2 can be theaverage of an appropriately large number of reshufflings of the state of Assumption 1. One particular characteristics of Assumption 2 is that the distancebetween each pair of minority class point is the same, therefore there does not exist any minority class point that will be examined "later" than otherswith respect to the number majority class points examined $\textbf{in sequence}\ $ before them.The problem here is that this assumption, even though it's closer to reality, is even worse for the Perceptron. In effect, with Assumption 1, despite the fact that we ended upwith a decision boundary tuned entirely with respect to one of the two classes (in fact, the class cardinalities have little to do in this case, it was the orderof training point examination that introduced the bias), we were able to correctly classify a portion of both classes on the training data (which might not be oureventual goal, but still looks good when we examine our algorithm). In this case, the minority class points will almost surely have no impact whatsoever inthe formation of the decision boundary. Unless we somehow tune the update rule of the Perceptron so that whenever we see a point of the minority class the weight isshifted a number of times the normal shifting (which is discussed in CIML in the sub-sampling / over-sampling section), or unless the feature values of the positivepoints are for some reason freakishly large with respect to the majority class's test points (which introduces complications; why not train a decision tree in this case?),it is certain that the decision boundary will not be able to shift enough to account for any single one of the minority class training points.To solve this problem, we resorted to the implementation of an oversampling scheme, which can be found in the train() and tune() methods of our OVA class in the submitted source code. In effect, given a minority and a majority class, we repeat the minority class $floor(\frac{|size(minority) - size(majority)|}{size(minority)})\ $ times and then we add the majority class. Prior to providing this data to one of the K - different Perceptrons to train on, we reshuffle it, because we aim tomake the OVA class applicable to any different binary classifier, and it is not guaranteed that every classification algorithm implementation contains references to reshuffling in its descriptive literature.
